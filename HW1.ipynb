#### №1

$$
tr(AA^*) = (\|A||_F)^2\\
tr(AA^*) = \sum(\lambda_i),\ \lambda_{max} - максимальное\ собственное\ значение\\
tr(AA^*) \leq rank(AA^*) \cdot \lambda_{max}\\
rank(AA^*) = rank(A) \Rightarrow\\ ||A||_F^2 \leq rank(A) \cdot \lambda_{max}\\
||A||_F \leq \sqrt{rank(A)} \cdot ||A||_2
$$

#### №2

Если А - единичная матрица, то условие будет выполнено, т.к. 
$$
\lambda_{max} = 1 \ и \ ||A||_F = \sqrt{rank(A)} \Rightarrow ||A||_F = \sqrt{(rank(A))}||A||_2
$$

#### №3

$$
A = U\Sigma V^* \leftarrow SVD\\
||AB||_F = ||U\Sigma V^* B||, т.к.\ U - унитарна \Rightarrow ||AB||_F = ||\Sigma V^* B||_F\\
\sigma_{max} - максимальное \ сингулярное \ значение\\
||AB||_F \leq \sigma_{max}||V^*B||_F, т.к. V - унитарна\\
||AB||_F \leq \sigma_{max}||B||_F \Rightarrow ||AB||_F \leq ||A||_2||B||_F
$$



####  №4

$$
A = UDU^{-1} \leftarrow спектральное \ разложение\\
т.к. \ A - эрмитова \Rightarrow A = UDU^*\\
\Lambda - матрица \ состоящая \ из \ -1 \ и \ 1 \  на \ диагонали, зависящими \  от \ знака \ собственных \  значений \ \\ 
на \ D \Rightarrow D \ можно \ представить \ как \  \Sigma\Lambda,где \Sigma - диагональная \ матрица \\
с \ модулями \ собственных \ значений.\\
\Rightarrow A=U\Sigma\Lambda U^*\\
$$

$$
т.к. \ U - унитарная \ матрица \Rightarrow \Lambda U^* - тоже \ унитарна, т.к. \ \Lambda \ только \ изменит \ значения \ некоторых \\
строк \ на \ (-1)\ представим \ \Lambda U^* = V^*\\
A=U\Sigma V^* \leftarrow сингулярное\ разлоэение\ , где \ сингулярные\ числа\ явл-ся \\
абсолютными\ значениями \ собственных\ значений.
$$



#### №5

$$
A - ортогональна \Rightarrow AA^T=A^TA=E\\
E_{ij} = \Sigma_{k=1}^{k=n}a_{ik}a_{jk} \leftarrow т.к. AA^T\\
для \ i \neq j\ E_{ij} = 0\Rightarrow \ в \ одном \ столбце \ матрицы \ не \ может \ стоять\ больше\ 1го\ значения\ больше\ 0 \\
Аналогично \ рассмотрим\ A^TA=E, из\ аналогии\ следует, что\ в\ одной\ строке\ может\ быть\\
только\ одно\ значение, которое\ не \ равно\ 0.\\
Т.к. \ E_{ij}=1=\Sigma a_{ik}a_{ik}, т.к. \ в \ каждой\ строке\ и \ столбце\ может\ быть\ только\ один\ элемент\\
который \ не \ равен \ нулю \Rightarrow \ подходит\ только\ случай\ когда\ положительные\ значения \ принимают\\
только\ единицу \Rightarrow A - матрица\ перестановки.
$$



#### №6

$$
A^* = V \Sigma U^*\\
A = U\Sigma V^*\\
\begin{bmatrix}
V & 0 \\
0 & U \\
\end{bmatrix} \times
\begin{bmatrix}
\Sigma & 0 \\
0 & \Sigma \\
\end{bmatrix}
\times
\begin{bmatrix}
U^*& 0 \\
0 & V^* \\
\end{bmatrix}
\times
\begin{bmatrix}
0 & I \\
I & 0 \\
\end{bmatrix}
=
\begin{bmatrix}
0 & A^* \\
A & 0 \\
\end{bmatrix}
\\
т.к.\ все\ матрицы\ в \ произведении\ унитарны \ и \ произведение\ унитарных\ матриц\ даёт\\
унитрную \Rightarrow нашли\ SVD
$$

#### №8

#### №10

$$
Если \ A - верхнетреугольная\ матрица, то \ A^* - нижнетреугольная\\
т.к. A - унитарна, следовательно\ AA^* = I \Rightarrow необходимо, чтобы \ произведения\ строк\ друг \ на \ друга\\
давало\ ноль\ (кроме\ произведения\ строки\ на \ саму\ себя), следовательно\ нам\ необоходимо\ чтобы\ все\ элементы\\
матрицы\ принимали\ 0, кроме\ диагональных, но\ на\ диагонали\ должны\ стоять\ единицы\ т.к.\ произведение \ строки\\
на \ саму\ себя\ будет\ давать\ квадрат\ диагонального\ элемента \Rightarrow треугольная\ унитарная\ матрица - диагональна
$$
